```python

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

iris_data = pd.read_csv("data/iris.csv")
```

```python
"""
1. Load the iris data from a csv file into a dataframe and 
print the shape of the data, type of the data and first 3 rows.

"""

print("Shape of the data:")
print(iris_data.shape)
print("\nData Type:")
print(type(iris_data))
print("\nFirst 3 rows:")
print(iris_data.head(3))
```

```python
"""
2. Print the keys, number of rows-columns, 
feature names and the description of the Iris data.
"""

print("\nKeys of Iris dataset:")
print(iris_data.keys())
print("\nNumber of rows and columns of Iris dataset:")
print(iris_data.shape) 
```

```python
"""
3. Display basic statistical details like percentile, mean, std etc. of iris data.

"""

print(iris_data.describe())
```

```python
"""
4. Create a Bar plot to get the frequency of the three species of the Iris data.
"""
ax=plt.subplots(1,1,figsize=(10,8))
sns.countplot('species',data=iris_data)
plt.title("Iris Species Count")
plt.show()
```

```python
"""
5. Create a graph to find relationship between the sepal length and width.
"""

fig = iris_data[iris_data.species=='setosa'].plot(kind='scatter',x='sepal_length',y='sepal_width',color='orange', label='Setosa')
iris_data[iris_data.species=='versicolor'].plot(kind='scatter',x='sepal_length',y='sepal_width',color='blue', label='versicolor',ax=fig)
iris_data[iris_data.species=='virginica'].plot(kind='scatter',x='sepal_length',y='sepal_width',color='green', label='virginica', ax=fig)

fig.set_xlabel("Sepal Length")
fig.set_ylabel("Sepal Width")
fig.set_title("Sepal Length VS Width")
fig=plt.gcf()
fig.set_size_inches(12,8)
plt.show()
```

```python
"""
6. Create a graph to see how the length and width of 
sepal_length, sepal_width, petal_length, petal_width are distributed.
"""

iris_data.hist(edgecolor='black', linewidth=1.2)
fig=plt.gcf()
fig.set_size_inches(12,12)
plt.show()
```

```python
"""
7. Create a joinplot to describe individual distributions 
on the same plot between Sepal length and Sepal width.
=========================================================

NB: 
1. The bivariate analogue of a histogram is known as a hexbin plot
2. It shows the counts of observations that fall within hexagonal bins. 
3. This plot works best with relatively large datasets. 
4. Its available through the matplotlib plt.hexbin function and as a style in jointplot(). 
5. It looks best with a white background.

"""

fig=sns.jointplot(x='sepal_length', y='sepal_width', kind="hex", color="red", data=iris_data)
plt.show()


# NB: it will be part of your homwwork to describe what this diagram represents. 

```
